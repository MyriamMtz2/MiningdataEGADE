```python
import numpy as np
import pandas as pd
```

```python
x={'company':['Ford','Ford','VW','Audi','Audi','Ford'],
     'price':[2,3,1,2,3,1],'weight':[4,6,2,2,4,3],'length':[6,7,3,6,8,np.nan],'made':['China','Mexico','Mexico','Corea','Uk','Uk']}
```

```python
df=pd.DataFrame(data=x);df
```

```python
df.isnull().values.any()
```

```python
df['weight'].isnull().values.any()
```

```python
df['length'].fillna(value = df['length'].mean(),
                    inplace = True
                   )
df
```

## what is the average of audi cars made in china?

```python
pd.pivot_table(df,
               index = 'company',
               values = 'price',
               aggfunc = 'sum'
              )
```

```python
pd.pivot_table(df,
               index = 'company',
               values = 'price',
               aggfunc = 'sum'
              )
```

```python
pd.pivot_table(df,
               index = 'company',
               values = 'price',
               aggfunc = ['sum','mean'],
               columns = 'made',
               margins = True,
               margins_name = '||**GrandTotal**||',
              )
```

```python
pd.pivot_table(df,
               index = 'company',
               values = ['price','weight'],
               aggfunc = ['sum','mean'],
               columns = 'made',
               margins = True,
               margins_name = '||**GrandTotal**||',
              )
```

```python
pd.pivot_table(df,
               index = 'company',
               values = ['price','weight'],
               aggfunc = {'price':np.mean,
                          'weight':np.sum},
               columns = 'made',
               margins = True,
               margins_name = '||**GrandTotal**||',
              )
```
