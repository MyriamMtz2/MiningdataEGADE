```python
df=pd.read_csv('datasets/antropometria-dataset.csv',
                 low_memory=False
                )
df.head(2)
```

```python
## pivot table sacando peso promedio
```

```python
pd.pivot_table(df,
               index = 'desc_ent',
               values = 'peso',
               aggfunc = 'mean'
              )
```

```python
## pivot table sacando mediana del peso 
```
