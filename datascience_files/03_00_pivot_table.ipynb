# Pivot Tables

```python
import numpy as np
import pandas as pd
df = pd.read_csv('datasets/census.csv')
#pd.set_option('display.max_columns',500, 'display.max_rows', 500)
df.head()
```

```python
pd.pivot_table(df, 
               index='STNAME', 
               values='BIRTHS2010', 
               aggfunc='mean'#'sum', 'median'
              )
```

```python
a = pd.pivot_table(df, 
               index='STNAME', 
               values='BIRTHS2010', 
               aggfunc='median'
               #columns='COUNTY'
              )
a.head()
```

```python
b = pd.pivot_table(df, 
               index='STNAME', 
               values=['BIRTHS2010','BIRTHS2011'], 
               columns='SUMLEV', 
               aggfunc = 'median',
               margins=True
              ) #margins gives the 'ALL' at the end and botom 
b.head()
```

```python
df = pd.read_csv('datasets/airline_data.csv',  low_memory=False)
df.head(3)
```

```python
a = df.columns.tolist()
a.sort()
a
```

```python
test1 = df.loc[:, ['DepDelay','DepartureDelayGroups']]
test2 = df['DepartureDelayGroups'].min()
test3 = df['DepartureDelayGroups'].max()
test4 = df.groupby(['DepDelay','DepartureDelayGroups']).size()
print(test2, test3)
print(test1)
print(test4)
```

```python
pd.pivot_table(df, 
               index=['Year', 'Quarter'], 
               values='DepDelay', 
               columns='DepartureDelayGroups', 
               aggfunc= 'sum',
               margins=True
              ) #margins gives the 'ALL' at the end and botom 
```

```python
pd.pivot_table(df, 
               index=['Year', 'Quarter'], 
               values='AirTime', 
               aggfunc='sum', 
               margins=True
              ) #margins gives the 'ALL' at the end and botom 
```

```python
data_file = 'datasets/sample_pivot.csv'
data = pd.read_csv(data_file)
data.head()
```

```python
pd.pivot_table(data, 
               index='Region'
              )
```

```python
average_sales = pd.pivot_table(data, 
                               index='Region', 
                               values='Sales'
                              )
average_sales
```

```python
total_sales_pivot = pd.pivot_table(data, 
                                   index='Region',
                                   values='Sales', 
                                   aggfunc='sum'
                                  )
total_sales_pivot
```

```python
sales_pivot = pd.pivot_table(data, 
                             index='Region', 
                             values='Sales', 
                             aggfunc=['sum', 'mean','median']
                            )
sales_pivot
```

```python
pd.pivot_table(data, 
               index='Region', 
               values='Units', 
               columns='Type'
              )
```

```python
pd.pivot_table(data, 
               index='Region', 
               values='Units', 
               columns='Type', 
               fill_value=0
              )
```

```python
pd.pivot_table(data, 
               index='Region', 
               values='Sales', 
               columns='Type', 
               aggfunc='sum', 
               margins='sum', 
               margins_name='Total'
              )
```
